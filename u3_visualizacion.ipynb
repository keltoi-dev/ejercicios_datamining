```python
import numpy as np
import pandas as pd
import datetime
from datetime import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1) Serie de datos
# Ejemplo 1
import numpy as np
import pandas as pd

lista=[
    [0, 1, 2, 3, 4], 
    [0, 1, 4, 9, 16]
]
 
df = pd.DataFrame(lista[1], index=lista[0])
df
```

```python
df.plot(title="Cuadrados")
plt.xlabel('Datos Y')
plt.ylabel('Datos x')
```

```python
# Ejemplo 2
np.random.seed(111)

s = pd.Series(np.random.randn(11), index=pd.date_range('2024-01-01', '2024-01-11'))
serie1 = s.cumsum()
serie1
```

```python
serie1.plot()
```

```python
df2 = pd.DataFrame(np.random.randn(11, 3), index=serie1.index, columns=list('ABC'))
df2 = df2.cumsum()
# df2.head()
df2
```

```python
df2.plot(title="Título")
plt.xlabel('Fechas')
plt.ylabel('Dato que varía')
```

```python
ax=df2.plot(title="Título")
ax.legend(['A', 'B', 'C'], loc='upper right')
```

```python
ax=df2.plot(title="Título", style=['b', '#9DD5FA', '#578EFA'])
ax.legend(['A', 'B', 'C'], loc="best")
```

```python
# Marcas
ax=df2.plot(title="Título", style=['b--', 'm--h', '#578EFA'], lw=2)
ax.legend(['A', 'B', 'C'], loc='upper right')
```

```python
# 3) Histograma
# Para una serie
df = pd.DataFrame(np.random.randn(1000))
df.hist()
```

```python
df.hist(bins=30)
```

```python
# Para varias series
df2 = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df2 = df2.cumsum()
df2.hist()
```

```python
# Para varias series superpuestas
x = pd.DataFrame(np.random.randn(1000))
y = pd.DataFrame(np.random.randn(1000))

bins = np.linspace(-10, 10, 100)
# alpha=0.5 50% transparente
plt.hist(x, bins, alpha=0.5, label='x')
plt.hist(y, bins, alpha=0.5, label='y')
plt.legend(loc='upper right')
```

```python
# 4) Diagrama de caja y bigotes
df2 = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df2 = df2.cumsum()
df2.boxplot()
```

```python
# 5) Dispersiones
df2 = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df2 = df2.cumsum()
df2.plot(kind='scatter', x='a', y='b')
```
