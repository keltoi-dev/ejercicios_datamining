<h3>PASO 1 - Importamos las librerias</h3>

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

<h3>PASO 2 - Cargamos el dataframe</h3>

```python
dataframe = pd.read_csv("student_performance.csv", encoding="ISO-8859-1")
dataframe
```

```python
# Creamos una copia para trabajar
dataset = dataframe.copy()
dataset.head()
```

<h3>PASO 3 - Limpieza de datos</h3>


```python
dataset = dataset.dropna()
dataset
```

<h3>PASO 4 - Convertimos la columna 'Extracurricular Activities'</h3>

```python
# Pasamos de datos categoricos (yes/no) a datos numericos
original_ea = dataset.pop("Extracurricular Activities")
dataset["Activities Extra"] = (original_ea == "Yes") * 1.0
dataset.head()
```

<h3>PASO 5 - Divido en entrenamiento y test</h3>

```python
# Separo los datos de entrada del de salida

data_x = dataset.copy()
data_y = dataset.pop("Performance Index")
data_x.pop("Performance Index")
```

```python
x_entrenar, x_test, y_entrenar, y_test=train_test_split(data_x, data_y)
print(len(x_entrenar))
print(len(x_test))
print("Porcentaje de datos de entrenamiento: ", len(x_entrenar)*100/(len(x_entrenar)+len(x_test)))
print("Porcentaje de datos de prueba: ", len(x_test)*100/(len(x_entrenar)+len(x_test)))
```

```python
sns.pairplot(x_entrenar[["Hours Studied", "Previous Scores", "Sleep Hours"]], diag_kind="kde")
```

```python
# Estadisticas generales
estadisticas = x_entrenar.describe()
estadisticas = estadisticas.transpose()
estadisticas
```

<h3>PASO 6 - Normalizado de datos</h3>

```python
def normalizar(valor):
	return (valor - estadisticas['mean']) / estadisticas['std']

entrenar_normal = normalizar(x_entrenar)
test_normal = normalizar(x_test)
entrenar_normal.head()
```
