```python
# shift() = Permite hacer desplazamientos
import pandas as pd 
ind = pd.date_range('01 / 01 / 2000', periods=5, freq='12H') 
df = pd.DataFrame({"A": [1, 2, 3, 4, 5], 
                "B": [10, 20, 30, 40, 50], 
                "C": [11, 22, 33, 44, 55], 
                "D": [12, 24, 51, 36, 2]}, 
                index=ind) 
print(df)
```

```python
df.shift(2, axis=0) 
```

```python
# cumprod() = Multiplica los valores de cada fila con los valores de la fila anterior
import pandas as pd

data = [[10, 18, 11], [13, 15, 8], [9, 20, 3]]

df = pd.DataFrame(data)
print(df)
print("---"*23)
print(df.cumprod())
```

```python
# rolling(t).mean() = saca el promedio sobre la cantidad de días (t) indicados
import pandas as pd 

ind = pd.date_range('01 / 01 / 2000', periods=5, freq='12h') 
df = pd.DataFrame({"A": [1, 2, 3, 4, 5], }, index=ind) 
print(df)
df.rolling(2).mean() 
```
